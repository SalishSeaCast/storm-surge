Notebook to plot the spatial extent of the surge over time. Dec 2006 case.

```python


%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

import arrow
import datetime
import pandas as pd
from salishsea_tools import tidetools
from salishsea_tools import stormtools_revisions
from salishsea_tools import nc_tools
from salishsea_tools import viz_tools

```

```python
fB = NC.Dataset('/data/nsoontie/MEOPAR/NEMO-forcing/grid/bathy_meter_SalishSea2.nc','r')

lats = fB.variables['nav_lat']
lons = fB.variables['nav_lon']
D = fB.variables['Bathymetry']
```

```python

#Load the data
path='/ocean/nsoontie/MEOPAR/SalishSea/results/storm-surges/revisions/dec2006/'

runs = {'all_forcing/restart', 'tidesonly/restart', 'ssh_only/restart'}

fUs={}; fVs={}; fTs={};

for key in runs:
    fUs[key] = NC.Dataset(path + key +'/SalishSea_1h_20061214_20061215_grid_U.nc','r');
    fVs[key] = NC.Dataset(path + key +'/SalishSea_1h_20061214_20061215_grid_V.nc','r');
    fTs[key] = NC.Dataset(path + key +'/SalishSea_1h_20061214_20061215_grid_T.nc','r');
    
    
```

```python
#Plot every two hours in a 3x4 structure.

ls=np.arange(29,39,1)
plot_ts= nc_tools.timestamp(fTs['all_forcing/restart'],ls)
plot_ts[0]
```

CGRF data

```python
CGRF = '/ocean/dlatorne/MEOPAR/CGRF/NEMO-atmos/'

U= NC.Dataset(CGRF +'u10_y2006m12d15.nc','r')
V= NC.Dataset(CGRF +'v10_y2006m12d15.nc','r')

u=U.variables['u_wind'][:]
v=V.variables['v_wind'][:]

tims=U.variables['time_counter'][:]

#need to truncrate CGRF domain
xs=454; xe=472
ys=518; ye=531

latU=U.variables['nav_lat'][:]
lonU=U.variables['nav_lon'][:]
lonU=lonU-360

latV=V.variables['nav_lat'][:]
lonV=V.variables['nav_lon'][:]
lonV=lonV-360

lat=latU[xs:xe,ys:ye]
lon=lonU[xs:xe,ys:ye]
u=u[:,xs:xe,ys:ye]
v=v[:,xs:xe,ys:ye]
```

```python
def plot_CGRF_wind(t, ax):
    #plots CGRF winds
    up = 0.5*(u[t,:,:] +u[t+1,:,:])
    vp = 0.5*(v[t,:,:] +v[t+1,:,:])
    
    quiver = ax.quiver(lon, lat, u[t,:,:], v[t,:,:],color='black',scale=45,pivot='mid', width=0.01)
    ax.quiverkey(quiver, -124.3, 48.9, 5, '5 m/s', coordinates='data', color='blue', labelcolor='blue')
```

```python
font = {'size'   : 10}

pylab.rc('font', **font)
```

```python
def plot_surge(run, plotCGRF):

    fig, axs = plt.subplots(2, 5, figsize=(25,25),sharey=True,sharex=True)
    depthlevel=0
    timesteps = ls
    vmin=0.35; vmax=.7
    cmap='Spectral_r'
    count=0
    for ax,t in zip (axs.flat,timesteps):
        Us={}; Vs={}; Es ={}; Ss={}; Ts={};

        for key in {run,'tidesonly/restart'}:
            [Us[key], Vs[key], Es[key], Ss[key], Ts[key]] = stormtools_revisions.get_variables(fUs[key],fVs[key],fTs[key],
                                                                                     t,depthlevel)
        
        mesh=ax.pcolormesh(lons[:],lats[:],Es[run]-Es['tidesonly/restart'],cmap=cmap,vmin=vmin,vmax=vmax)
        viz_tools.set_aspect(ax)
        viz_tools.plot_coastline(ax,fB,coords='map')
        ax.set_title(plot_ts[count].format('MM/DD HH:mm'))
        ax.set_xlim([-124.7,-122.5])
        ax.set_ylim([47.5,50])
        ax.set_xticks([-124,-123])
        count=count+1
        
    if (plotCGRF):
        CGRFtims =np.array(np.linspace(5,15,11),int)
        CGRF_ts= nc_tools.timestamp(U,CGRFtims)
        print CGRF_ts
        for ax,t in zip (axs.flat,CGRFtims):
            plot_CGRF_wind(t,ax)
    
    fig.set_size_inches(12,6)

    for ax in axs[1,:]:
        ax.set_xlabel('Longitude')
    for ax in axs[:,0]:
        ax.set_ylabel('Latitude')
    fig.subplots_adjust(hspace=-.1)
    fig.subplots_adjust(wspace=0.1)
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.82, 0.25, 0.02, 0.5])
    cbar=fig.colorbar(mesh, cax=cbar_ax)
    cbar.set_label('Surge (m)')
    
    return fig
```

```python
fig=plot_surge('all_forcing/restart',True)
```

```python
fig.savefig('../RoughWork/Figures/dec2006_spatial.png',bbox_inches='tight',dpi=300)
```
