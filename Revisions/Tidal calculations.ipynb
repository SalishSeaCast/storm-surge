Forcing residuals were calculated based on a tidal prediction using constituents from a t_tide analysis from the year prior to the surge event. I also used the signal to noise ratio cut off in t_tide. 

Based on revision history, it looks like snr > 10 was used in the forcing tidal predictions, but no cut off was used in the analysis at PA, Victoria, etc. (except perhaps the 2012 simulations).

This notebook will investigate this in detail by recalculating the tidal predictions and trying to decide which snr was used if any. 

In the cases where snr is applied, I will determine which constituents were exluded. 



```python
import pandas as pd
import matplotlib.pyplot as plt
import collections

%matplotlib inline
import numpy as np
```

```python
def analyze_snr(hfile,snr):
    """prints constituents in a harmonic analysis that have a signal to noise ration greater than snr"""
    
    data = pd.read_csv(hfile,delimiter='\t',skiprows=2)
    
    print 'Total number of constituents:', len(data.index)
    
    snr_filt = data[(data[' amp (m) ']/data[' amp error '])**2>snr]
    
    print 'Number of constituents with snr >', snr, ':', len(snr_filt.index)
    print 'Constituents with snr >', snr
    print snr_filt
```

```python
def interannual_variability_compare(files):
    fig,ax=plt.subplots(1,1,figsize=(20,10))
    
    #initialzie
    data = pd.read_csv(files[files.keys()[0]],delimiter='\t',skiprows=2)
    temp = {'Constituent': data['Constituent '].values}

    d=pd.DataFrame(temp)
    
    for year, hfile in files.iteritems():
        data=pd.read_csv(hfile,delimiter='\t',skiprows=2)
        
        d[year] = data[' amp (m) '].values
    #plotting
    d=d.set_index('Constituent')
    d.plot(kind='bar',ax=ax)
    
    #range
    print 'Constituents with > 1cm change in amplitude between years'
    r = d.max(axis=1,numeric_only=True)-d.min(axis=1,numeric_only=True)
    print r[r>0.01]
    plt.legend()
    return fig
```

# Forcing residuals

## Tofino

```python
files = collections.OrderedDict([
        ('2005', '/data/nsoontie/MEOPAR/tools/I_ForcingFiles/OBC/Tofino_harmonics_31-Dec-2004_02-Jan-2006.csv'),
        ('2008', '/data/nsoontie/MEOPAR/tools/I_ForcingFiles/OBC/Tofino_harmonics_31-Dec-2007_02-Jan-2009.csv'),
        ('2011', '/data/nsoontie/MEOPAR/tools/I_ForcingFiles/OBC/Tofino_harmonics_31-Dec-2010_02-Jan-2012.csv')
        ])

for year,hfile in files.iteritems():
    print year
    analyze_snr(hfile,10)

fig =interannual_variability_compare(files)
```

### Summary

Only 2006 used that Sa tidal constituent. SSa and MM were not used in any Tofino residuals. There is a lot of interannual variability in the long period constituents.

# Notes on Interannual Variability

If Sa, SSa, MM have an interannual variability then I don't think we should be using them in the calculation of the residuals. The idea is that interannual variability is captured in the residual and forced that way. Same goes for Z0. Any interannula variability in the mean sea level should be included in the residual calculation.

When I review the creation of the forcing residuals, I used that year's Z0 value in the tidal predictions. Why did I do that? Shouldn't I be using a long term mean sea level in order to capture the interannual variability?
