This notebook compare tidal predictions for 2006 at Tofino calculated with several different methods.

1. As in paper (2005 harmonics, SNR>10)
2. 2005 harmonics, SNR>2
3. 2006 harmonics, SNR>2
4. 2006 haromncs, storm surge removed
5. CHS haromnics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from dateutil import tz
import datetime
import pytz


from salishsea_tools import stormtools
from salishsea_tools import tidetools
from salishsea_tools.nowcast import figures

%matplotlib inline
```

```python
def dateParserMeasured2(s):

    PST=tz.tzoffset("PST",-28800)
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%d-%b-%Y %H:%M:%S ")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=PST)
    #convert to UTC
    return aware.astimezone(tz.tzutc())
```

```python
def load_tidal_predictions(filename):

    #read msl
    line_number = 1
    with open(filename, 'rb') as f:
        mycsv = csv.reader(f); mycsv = list(mycsv)
        msl = mycsv[line_number][1]
        msl=float(msl)

    ttide = pd.read_csv(filename,skiprows=3,parse_dates=[0],date_parser=dateParserMeasured2)
    ttide = ttide.rename(columns={'Time_Local ': 'time', ' pred_8 ': 'pred_8', ' pred_all ': 'pred_all'})

    return ttide, msl
```

```python
def plot_tides(key, tfile,ax,color='k'):
    
    tides, msl = load_tidal_predictions(tfile)
    
    ax.plot(tides.time,tides.pred_all,'-',c=color,label=key,lw=1.5)
```

```python
files = {
 '2005_snr10': 'Tofino_t_tide_compare8_31-Dec-2005_02-Jan-2007.csv',
 '2005_snr2': 'Tofino_t_tide_compare8_31-Dec-2005_02-Jan-2007_snr2.csv',
 '2006_snr2': 'Tofino_t_tide_compare8_31-Dec-2005_02-Jan-2007_snr2_self.csv',
 'CHS': 'Tofino_atide_compare8_31-Dec-2005_02-Jan-2007.csv'}

colors= {
 '2005_snr10':'r',
 '2005_snr2': 'b',
 '2006_snr2': 'g',
 'CHS': 'k'}
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))

for key,tfile in files.iteritems():
    plot_tides(key,tfile,ax,colors[key])

ax.set_xlim([datetime.datetime(2006,2,1),datetime.datetime(2006,2,7)])
plt.legend(loc=0)
ax.grid()
```

```python

def dateParserMeasured(s):
    PST_tz = tz.tzoffset("PST", -28800)
    # Convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%Y/%m/%d %H:%M")
    # Add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=PST_tz)
    # Convert to UTC
    return aware.astimezone(tz.tzutc())
```

```python
statID = '8615'#Tofino station ID number
s = '31-Dec-2005' 
e= '02-Jan-2007'

tidetools.get_dfo_wlev(statID,s,e)
filename ='wlev_' +str(statID) + '_' + s +'_' +e +'.csv'
    
obs = pd.read_csv(filename,skiprows=7,parse_dates=[0],date_parser=dateParserMeasured)
obs = obs.rename(columns={'Obs_date': 'time', 'SLEV(metres)': 'slev'})
```

```python
def calculate_residual(obs,ttide,msl):
    ssanomaly = np.zeros(len(obs.time))

    for i in np.arange(0,len(obs.time)):
    #check that there is a corresponding time 
    #if any(wlev_pred.time == wlev_meas.time[i]):
        ssanomaly[i] =(obs.slev[i] - (ttide.pred_all[ttide.time==obs.time[i]]+msl))
        if not(ssanomaly[i]):
            ssanomaly[i]=float('Nan')
    
    return ssanomaly
```

```python
def plot_residual(key,tfile,ax,color='k'):
    tides, msl = load_tidal_predictions(tfile)
    
    residual = calculate_residual(obs,tides,msl)
    
    ax.plot(obs.time,residual,'-',c=color,label=key,lw=1.5)
    
    return residual
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))

residuals={}

for key,tfile in files.iteritems():
    res=plot_residual(key,tfile,ax,colors[key])
    residuals[key]=res
ax.set_xlim([datetime.datetime(2006,2,1),datetime.datetime(2006,2,7)])
plt.legend(loc=0)
ax.grid()
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))


for key,res in residuals.iteritems():
    ax.plot(obs.time, res,c=colors[key],label=key,lw=1.5)
ax.set_xlim([datetime.datetime(2006,12,10),datetime.datetime(2006,12,17)])
plt.legend(loc=0)
ax.grid()
```

```python
def plot_power_spectrum(key, residual,ax,color='k'):
    t1=obs.time.iloc[0]
    t2=obs.time.iloc[-1]
    tim_diff = (t2-t1).total_seconds()/60.
    ts = [t1 + datetime.timedelta(hours=dt) for dt in np.arange(0,tim_diff)]
    
    
    res_inter=figures.interp_to_model_time(ts,residual,obs.time)
    # model
    ps = np.abs(np.fft.fft((res_inter)))**2
    freqs = np.fft.fftfreq(len(res_inter), 1)
    ax.plot(freqs,ps,'o',c=color,label=key)


```

```python
fig,axs=plt.subplots(4,1,figsize=(20,10))

for key,ax in zip(residuals,axs):
    plot_power_spectrum(key, residuals[key],ax,color=colors[key])
    ax.set_ylim([0,1000])
    ax.grid()
    ax.set_title(key)
```
